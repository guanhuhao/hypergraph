```python
def load_data(path):
    mp_node = {}
    mp_edge = {}
    with open(path,'r') as f:
        for line in f:
            n_id,e_id = line[0:-1].split(" ")
            if mp_node.get(n_id) == None : 
                mp_node[n_id] = 0
            if mp_edge.get(e_id) == None :
                mp_edge[e_id] = 0
                
            mp_node[n_id]+= 1
            mp_edge[e_id]+= 1
            
    Edges = []
    for edge in mp_edge.values():
        Edges.append(edge)
    return Edges,len(mp_node)

Edges,n = load_data("../data/github/github.txt")
print("finished load data!")
```

```python
from scipy.special import comb

def remove(Edges,n,p,x):
    # Edges: list of edge degree
    # n: number of vertex
    # p: number of partition 
    # x: number of remove node 
    
    np = n/p
    ret = 0
    cnt = 0;
    for edge in Edges:
        for j in range(1,edge+1):
            fac = comb(edge,j)*(1-1/p)**(edge-j) *(1/p)**j
            ret  += fac * ( comb(np,x-j) / comb(np,x) ) 
#             print(cnt)
#             cnt+=1
    return ret;

def add(Edges,n,p):
    ret = 0;
    for edge in Edges:
        ret += (edge/n)*((1-1/p)**(edge-1))
    return ret

def func(Edges,n,p,remove_degree,k_old):
    cost = 0;
    remove_node = remove_degree/k_old
    
    gain1 = remove(Edges,n,p,remove_node)
    cost += gain1
    
    gain2 = 0;
    for i in range(1,remove_degree-k_old+2):
        gain2 += (k_old-1) * (comb(remove_degree-1-i,k_old-1-i) / comb(remove_degree,k_old-1))
    cost += gain2
        
    cost -= 2 * remove_node * add(Edges,n,p)
    cost += (remove_degree/n)*((1-1/p)**(remove_degree-1))
    
#     print("remove1:",remove_node," gain:",remove(Edges,n,p,remove_node));
#     print("remove2:",remove_node/k_old," gain:",remove(Edges,n,p,remove_node/k_old));
    print("gain1:",round(gain1,2)," gain2:",round(gain2)," loss:",round(2 * remove_node * add(Edges,n,p),2),end =" ")
    print("fix:",round((remove_degree/n)*((1-1/p)**(remove_degree-1))))
    return cost
    
    
# Edges = []
# for i in range(10):
#     Edges.append(i)
# n = 100
p = 4
x = 2
# print(remove(Edges,n,p,2))
# print(add(Edges,n,p))
for i in range(1,10):
    print("i:",i,"cost:",func(Edges,n,p,i+1,2))


    
```

```python
# Setup
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
from scipy.spatial import Delaunay

params = {'font.size'     : 14,
          'figure.figsize':(15.0, 8.0),
          'lines.linewidth': 2.,
          'lines.markersize': 15,}
matplotlib.rcParams.update(params)

Ni = 40
Pi = np.random.rand(Ni, 2)
Xi, Yi = Pi[:,0], Pi[:,1]
Zi = np.random.rand(Ni)

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot(Xi, Yi, Zi, "or", label='Data')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()

Pi = np.array([Xi, Yi]).transpose()
tri = Delaunay(Pi)
plt.triplot(Xi, Yi , tri.simplices.copy())
plt.plot(Xi, Yi, "or", label = "Data")
plt.grid()
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()

N = 100
x = np.linspace(0., 1., N)
y = np.linspace(0., 1., N)
X, Y = np.meshgrid(x, y)
P = np.array([X.flatten(), Y.flatten() ]).transpose()
plt.plot(Xi, Yi, "or", label = "Data")
plt.triplot(Xi, Yi , tri.simplices.copy())
plt.plot(X.flatten(), Y.flatten(), "g,", label = "Z = ?")
plt.legend()
plt.grid()
plt.show()

Z_cubic = griddata(Pi, Zi, P, method = "cubic").reshape([N, N])
plt.contourf(X, Y, Z_cubic, 50, cmap = mpl.cm.jet)
plt.colorbar()
plt.contour(X, Y, Z_cubic, 20, colors = "k")
#plt.triplot(Xi, Yi , tri.simplices.copy(), color = "k")
plt.plot(Xi, Yi, "or", label = "Data")
plt.legend()
plt.grid()
plt.show()
```
