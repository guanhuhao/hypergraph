```python
import matplotlib.pyplot as plt 
# path = "./github.txt.log"
# path = "./out.actor-movie.log"
# path = "./out.dblp-author.log"
path = "./out.dbpedia-location.log"
# path = "./out.dbpedia-team.log"
# path = "./wiki.txt.log"

with open(path,"r") as file:
    data_p = {}
    for line in file:
        if line[0] == "#" : continue
        p,sheild,per,k_1,runtime = line[0:-1].split(" ")
        p = int(p)
        sheild = int(sheild)
        per = float(per)
        k_1 = int(k_1)
        runtime = int(runtime)
        
        if data_p.get(p) == None : data_p[p] = []
        
        data_p[p].append((sheild,per,k_1,runtime))
        
def draw_per_k(p):
    x = [item[1] for item in data_p[p]]
    y = [item[2] for item in data_p[p]]
    print(x)
    print(y)
#     plt.yscale('log')
    plt.xscale('logit')
    plt.plot(x,y)
    plt.show()
    
    rec_pos = 0  
    for i in range(len(y)): 
        if y[i] < y[rec_pos]: 
            rec_pos = i
    print("sheild per:",x[i],"mini k-1:",y[i])
    
def draw_per_time(p):
    x = [item[1] for item in data_p[p]]
    y = [item[3] for item in data_p[p]]
    print(x)
    print(y)
#     plt.yscale('log')
    plt.xscale('logit')
    plt.plot(x,y)
    plt.show()
    
    rec_pos = 0  
    for i in range(len(y)): 
        if y[i] < y[rec_pos]: 
            rec_pos = i
    print("sheild per:",x[i],"mini k-1:",y[i])
    
draw_per_k(8)
draw_per_time(8)
    
    
```
