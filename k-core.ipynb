```python
path = "./data/github/vertex_stream.txt"
class Node:
    def __init__(self,id,degree,edge_list):
        self.id = id
        self.degree = degree 
        self.edge_list = edge_list

mp = {}
bia = 1000000000

with open(path,'r') as f:
    for line in f:
        if(line[0]=='#' or line=='\n') : continue
        data = [int(i) for i in line[0:-1].split(" ")]
        n_id = data[0]
        n_degree = data[1]
        n_edges = set(data[2:])
        if mp.get(n_id) == None : mp[n_id] = Node(n_id,0,[])
        for edge in n_edges:
            e_id = edge + bia;
            if mp.get(e_id) == None : mp[e_id] = Node(e_id,0,[])
            mp[n_id].degree += 1
            mp[n_id].edge_list.append(e_id)
            
            mp[e_id].degree += 1
            mp[e_id].edge_list.append(n_id)
op = {key:value for key,value in mp.items()}



k = 0

while len(op) > 10:
    degree = {}
    for node in op.values():
        if degree.get(node.degree) == None : degree[node.degree] = []
        degree[node.degree].append(node)
    degree = [(key,value) for key,value in degree.items()]
    degree.sort(key = lambda x:x[0])
    
#     print([(i,len(j)) for i,j in degree])
    for key,value in degree:
        if key > k : break
        for node in value:
            for edge in node.edge_list:
                if op.get(edge) != None:
                    op[edge].degree -= 1
            del op[node.id]
    k += 1
    print("k:"+str(k) +"   ret:"+str(len(op)))
            
```
