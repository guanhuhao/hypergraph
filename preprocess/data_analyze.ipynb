```python
import matplotlib.pyplot as plt # 画曲线图
import numpy as np #拟合直线
def load_data(filename):
    n_degree = {}
    e_degree = {}
    with open(filename,'r') as file:
        for line in file:
#             print( line[0:-1].split(" ")[0:2])
            n_id,e_id = line[0:-1].split(" ")[0:2]
            if n_degree.get(n_id) == None: n_degree[n_id] = 0;
            if e_degree.get(e_id) == None: e_degree[e_id] = 0;
            n_degree[n_id] += 1
            e_degree[e_id] += 1
    return n_degree,e_degree
load_data(filename);

def degree_distribute(degree,show = True,logx=True,logy=True,coeff = True,cumulation = False): # draw degree distribute graph 
    dic = {j:0 for i,j in degree.items()}
    
    for i,j in degree.items(): dic[j] += 1
    
    ls = sorted(dic.items(), key = lambda x:(x[0],x[1]))
    degree = sorted(list(degree.items()),key = lambda x:(x[1],x[0]))
        
    x = [i[0] for i in ls]
    y = [i[1] for i in ls]

    if x[0] == 0:
        x = x[1:-1]
        y = y[1:-1]
    
    if cumulation == True : 
        cnt = sum(y)
        for i in range(len(x)):
            cnt -= y[i]
            y[i] += cnt

    if show == True :
        plt.plot(x,y, 'o',color='b')
        if logy : plt.yscale('log')
        if logx : plt.xscale('log')
        

    if coeff == True : 

        xx = [np.log2(i) for i in x]
        yy = [np.log2(i) for i in y]
        
        k,b = np.polyfit(xx, yy, 1)
#         lg = yy[0]/xx[-1]
#         lg = 1/lg
        print("alpha: "+str(k))
        
        xx = x
        yy = [(i**k)*(2**b) for i in xx]
        plt.plot(xx,yy,color='r')
    plt.show()

    return [(x[i],y[i]) for i in range(len(x))]

def ask(ls,val):
    cnt = 0;
    for i in ls:
        cnt += 1
        if i >= val : break
    return 1.0*cnt/len(ls)
```
